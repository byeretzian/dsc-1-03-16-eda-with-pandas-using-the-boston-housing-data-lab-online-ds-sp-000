# subsets

### Charles River Adjacent vs. Not

##### I am interested in whether the median value of homes on the Charles River is greater than those that are not. I believe that the mean and median of the home prices will be greater, but the standard deviation will be less.

```python
#creating the subsets
CRA = df.loc[df['chas'] == 1]
NCRA = df.loc[df['chas'] == 0]
```

```python
# analyzing the descriptive statistics 
CRA.describe() # mean: 30.175, std:12.362204, median: 26.05, IQR: (37.9 - 21.7) = 16.2
NCRA.describe() #22.295527, std: 8.746397, median: 21.2, IQR: (24.7 - 17.2) = 7.5
# the mean and median prices for homes on the river is significantly higher than the prices of other homes.
```

```python
# graphing data
fig = plt.figure(figsize = (12,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
CRA_plot = CRA.hist(column = 'medv', ax = ax1)
NCRA_plot = NCRA.hist(column = 'medv', ax=ax2)
# the histograms show that most homes not on the river are priced between 10-30, with barely any priced above. Homes on the river average about 22-27, with many of them valued at 45-50.
```

### African American Population >= Median vs Not

##### I am interested to see if the pupil - teacher ratio is different in towns with higher proportions of African Americans. I can think of no reason for this to be the case, so I don't think the measures of centrality and dispersion of the two dataframes will be very different.

```python
# creating the subsets
MAA = df.loc[df['black'] >= 392.05]
NMAA = df.loc[df['black'] < 392.05]
```

```python
# analyzing the descriptive statistics
MAA.describe() # mean: 18.486826, std: 1.570994, median: 18.6, IQR: (20.2 - 17.7) = 2.5
NMAA.describe() # mean: 18.409036, std: 2.319191, median: 19.1, IQR: (20.2 - 16.925) = 3.275
# As predicted, the statistics for the pupil to teacher ratio in areas with higher populations of African Americans is not very different.
```

```python
# graphing data
fig = plt.figure(figsize = (12,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
MAA_plot = MAA.hist(column = 'ptratio', ax = ax1)
NMAA_plot = NMAA.hist(column = 'ptratio', ax=ax2)

#the graphs show almost identical histograms
```

### Nitrogen Oxide Concentration >= Mean vs Not

##### I would like to compare the air polution in areas that are closer to Boston's employment centers to areas that are farther, and I believe proximity to these locations is related to higher levels of nitrogen oxides in the air.

```python
# creating the subsets
NOCM = df.loc[df['nox'] >= 0.557144]
NOCL = df.loc[df['nox'] < 0.557144]
```

```python
# analyzing the descriptive statistics
NOCM.describe() # mean: 2.052759, std: 0.492744, median: 1.996350
NOCL.describe() # mean: 4.912030, std: 1.776492, median: 4.502600
# the mean and median distance from employment centers for areas with less than the mean concentration of nitrogren oxide are more than double those of areas with higher concentration.
```

```python
# graphing data
fig = plt.figure(figsize = (14,6))
ax1 = fig.add_subplot(121)
ax1.set_xlim(0, 12), ax1.set_ylim(0,1)
ax2 = fig.add_subplot(122)
ax2.set_xlim(0, 12), ax2.set_ylim(0,1)
NOCM_plot = NOCM.plot.scatter( 'dis', 'nox',  ax=ax1, title = 'Closer')
NOCL_plot = NOCL.plot.scatter('dis', 'nox', ax=ax2, title = 'Farther')
# the plots show that all of of the points with higher than average pollution are within 4 units of distance from city centers. The points with lower than average pollution tend to be farther away.                
```

# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.
